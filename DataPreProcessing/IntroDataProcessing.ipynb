## EDA & Outliers

```python
num_cols = df.select_dtypes(include=["float64", "int64"]).columns.tolist()
col_outliers = list()

for feature in num_cols:

    # Histogram
    ax = sns.histplot(df[feature], kde=True)
    ax.set_title(f"Distribución de {feature}")
    ax.set_xlabel("Ventas")
    ax.set_ylabel("Frecuencia")

    # Boxplot -> Visual method
    plt.figure(figsize=(8, 6))
    ax = sns.boxplot(y=feature, data=df)
    ax.set_title(f"Boxplot de {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("Distribución")

    # Plot
    plt.show()

    # Outliers
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers = df[(df[feature] < limite_inferior) | (df[feature] > limite_superior)]

    # Save outliers
    if outliers.empty:
        print("No hay outliers\n")
    else:
        col_outliers.append(feature)
        print(f"Outliers en {feature}: {outliers.index}\n")
        print(outliers)

print(f"Features with outliers: {col_outliers}")
```

## Imputation

```python
# Imput Missing Values
from sklearn.impute import SimpleImputer

# Imputer mean
imputer_mean = SimpleImputer(strategy="mean")
df['Edad'] = imputer_mean.fit_transform(df[['Edad']])
imputer_mean = SimpleImputer(strategy="mean")
df['Calificación_satisfacción'] = imputer_mean.fit_transform(df[['Calificación_satisfacción']])
imputer_mean = SimpleImputer(strategy="mean")
df['Descuento_promedio'] = imputer_mean.fit_transform(df[['Descuento_promedio']])

# Imputer most_frequent
imputer_most_frequent = SimpleImputer(strategy="most_frequent")
df["Activo"] = imputer_most_frequent.fit_transform(df[["Activo"]]).ravel()
imputer_most_frequent = SimpleImputer(strategy="most_frequent")

# Check Missing Values
df.isnull().mean()*100
```

```python
# Get Categorical Features
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
print(cat_cols)
```

```python
# Encode Features
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# Encode Pais
OneHot = OneHotEncoder(sparse_output=False, drop="first")
pais_encoded = OneHot.fit_transform(df[["País"]])
df = df.drop(columns=["País"])
pais_columnas = OneHot.get_feature_names_out(["País"])
df[pais_columnas] = pais_encoded

# Encode Producto_frecuente
OneHot = OneHotEncoder(sparse_output=False, drop="first")
Producto_frecuente_encoded = OneHot.fit_transform(df[["Producto_frecuente"]])
df = df.drop(columns=["Producto_frecuente"])
Producto_frecuente_columnas = OneHot.get_feature_names_out(["Producto_frecuente"])
df[Producto_frecuente_columnas] = Producto_frecuente_encoded

# Rango_crediticio
ord_cat = ["Bajo", "Medio", "Alto", "Premium"]
orden_nivel = OrdinalEncoder(categories=[ord_cat])
df["Rango_crediticio"] = orden_nivel.fit_transform(df[["Rango_crediticio"]])

# Check
df.head()
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler

# Columns
scaling_columns = ["Edad", 'Ingreso_anual', 'Frecuencia_compra', 'Calificación_satisfacción', 'Descuento_promedio', 'Días_desde_última_compra']

# Standard Scaling
scaler = StandardScaler()
df[scaling_columns] = scaler.fit_transform(df[scaling_columns])

# Check
df.head()
```
