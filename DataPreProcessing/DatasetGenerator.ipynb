```python
import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker()
np.random.seed(123)

# Número de registros que quieres generar
n = 10  # Puedes poner 100, 1000, etc.

# Generar datos simulados

nuevo_data = {
    'ID_cliente': [f'N{2000+i}' for i in range(n)],
    'Nombre_cliente': [fake.name() for _ in range(n)],
    'Edad': np.random.choice(list(range(18, 80)) + [np.nan], n),
    'Género': np.random.choice(['M', 'F', 'Otro', np.nan], n),
    'Estado_civil': np.random.choice(['Soltero', 'Casado', 'Viudo', 'Divorciado', np.nan], n),
    'Ingreso_anual': np.random.choice(list(range(10000, 200000)) + [np.nan], n),
    'Número_hijos': np.random.choice([0, 1, 2, 3, 4, np.nan], n),
    'País': np.random.choice(['México', 'Colombia', 'Argentina', 'Chile'], n),
    'Fecha_registro': [fake.date_between(start_date='-5y', end_date='today') for _ in range(n)],
    'Frecuencia_compra': np.random.choice(
        list(np.round(np.random.uniform(0, 20, n), 1)) + [np.nan], n),
    'Calificación_satisfacción': np.random.choice(
        list(np.round(np.random.uniform(0, 10, n), 1)) + [np.nan], n),
    'Producto_frecuente': np.random.choice(['Electrónica', 'Ropa', 'Hogar', 'Deportes', 'Belleza'], n),
    'Email': [fake.email() for _ in range(n)],
    'Descuento_promedio': np.random.choice(list(range(0, 51)) + [np.nan], n),
    'Activo': np.random.choice([True, False, np.nan], n),
    'Código_postal': [fake.postcode() for _ in range(n)],
    'Comentarios_cliente': [fake.sentence() for _ in range(n)],
    'Días_desde_última_compra': np.random.choice(list(range(0, 1825)) + [np.nan], n),
    'Rango_crediticio': np.random.choice(['Bajo', 'Medio', 'Alto', 'Premium'], n),
    'Comprará_de_nuevo': np.random.choice([0, 1], n)  # Target binario
}

nuevo_dataset = pd.DataFrame(nuevo_data)

# Guardar como CSV
#nuevo_dataset.to_csv('nuevo_dataset.csv', index=False)
```
