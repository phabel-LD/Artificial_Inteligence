# UMAP

```python
from sklearn.datasets import fetch_openml
import numpy as np
import matplotlib.pyplot as plt

mnist = fetch_openml('mnist_784', version=1)

X = np.array(mnist.data)
y = np.array(mnist.target)

X = X[:10000]
y = y[:10000]


print("Forma de X:", X.shape)
print("Forma de y:", y.shape)
```

```python
plt.figure(figsize=(10, 10))

n_images = 10
indices = np.random.choice(X.shape[0], n_images, replace=False)

for i, index in enumerate(indices):
    # Obtener la imagen y la etiqueta
    image = X[index].reshape(28, 28)  # Redimensionar a 28x28
    label = y[index]

    # Crear subgr치fica
    plt.subplot(5, 5, i + 1)
    plt.imshow(image, cmap='gray')
    plt.title(f'Label: {label}')
    plt.axis('off')  # Ocultar ejes

# Mostrar la figura
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y.astype(int), cmap='viridis', alpha=0.5, s=5)
plt.title('PCA de MNIST')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(scatter, label='Etiqueta')
plt.show()
```

```python
import umap
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
import numpy as np

umap_model = umap.UMAP(n_components=2, n_neighbors=15, min_dist=0.1, random_state=42)
X_umap = umap_model.fit_transform(X)


plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y.astype(int), cmap='viridis', alpha=0.5, s=5)
plt.title('UMAP de MNIST')
plt.xlabel('Dimensi칩n 1')
plt.ylabel('Dimensi칩n 2')
plt.colorbar(scatter, label='Etiqueta')
plt.show()
```
