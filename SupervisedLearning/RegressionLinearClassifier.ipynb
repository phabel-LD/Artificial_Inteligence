# RegresiÃ³n LogÃ­stica

````python
tabla = pd.read_csv("https://gitlab.com/dgtic5/res/-/raw/main/aprendizajeSupervizado/Diabetes__Edited_.csv")
````

````python
tabla
````

````python
y = tabla["I"]
X = tabla["B"].to_numpy().reshape(-1,1) # Se debe indicar explÃ­citamente que es una matriz de una sola columna, en lugar de un vector.
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
````

````python
from sklearn.linear_model import LogisticRegression
````

````python
modelo = LogisticRegression()
modelo.fit(X_train,y_train)
````

````python
modelo.coef_
````

````python
y_pred = modelo.predict(X_test)
````

````python
y_pred
````

## Visualization

````python
import matplotlib.pyplot as plt
import numpy as np

x_func = np.linspace(0,200,1000)
y_func = modelo.predict(x_func.reshape(-1, 1))
x_points = X_train
y_points = y_train

fig, ax = plt.subplots()
ax.plot(x_func,y_func)
ax.scatter(x_points,y_points,s=60)
````

````python
y_prob = modelo.predict_proba(x_func.reshape(-1, 1))

fig, ax = plt.subplots()
ax.plot(x_func,y_func)
ax.plot(x_func,y_prob)
ax.scatter(x_points,y_points,s=60)
````

# Multinomial

````python
from sklearn.datasets import load_iris
````

````python
dataset = load_iris(as_frame=True)
X = dataset.data
y = dataset.target
````

````python
X
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
````

````python
modelo = LogisticRegression()
````

````python
modelo.fit(X_train,y_train)
````

````python
y_pred = modelo.predict(X_test)
````

## Evaluation: MICRO y MACRO

````python
micro = f1_score(y_test,y_pred,average="micro")
macro = f1_score(y_test,y_pred,average="macro")
weighted = f1_score(y_test,y_pred,average="weighted")

print(macro)
print(weighted)
print(micro)

print(accuracy_score(y_test,y_pred))
````

## Classification Report

````python
from sklearn.metrics import classification_report
````

````python
print(classification_report(y_test,y_pred))
````

# Multi-Label

````python
from sklearn.datasets import make_multilabel_classification
````

````python
X, y = make_multilabel_classification(n_samples=100,n_features=10, n_classes=5, n_labels=3, random_state=2)
````

````python
y
````

````python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
````

````python
from sklearn.multiclass import OneVsRestClassifier

modelo = OneVsRestClassifier(LogisticRegression())
````

````python
modelo.fit(X_train,y_train)
````

````python
y_pred = modelo.predict(X_test)
````

````python
micro = f1_score(y_test,y_pred,average="micro")
macro = f1_score(y_test,y_pred,average="macro")
weighted = f1_score(y_test,y_pred,average="weighted")
samples = f1_score(y_test,y_pred,average="samples")

print(macro)
print(weighted)
print(micro)
print(samples)

print(accuracy_score(y_test,y_pred))
````

````python
print(classification_report(y_test,y_pred))
````
