```python
# Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import accuracy_score, f1_score, classification_report
```

```python
# Load Data
table = pd.read_csv("https://gitlab.com/dgtic5/res//raw/main/aprendizajeSupervizado/column_3C_weka.csv")
table
```

```python
# Set Labels
X = table.drop("class", axis=1)
y = table["class"]
```

```python
# Training Data
X
```

```python
# Target Labels
y.value_counts()
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
# Model with integrated Cross Validation
model = LogisticRegressionCV(cv=5, random_state=0, max_iter=50).fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
# Micro-Macro-Weighted F1-Scores
micro = f1_score(y_test,y_pred,average="micro")
macro = f1_score(y_test,y_pred,average="macro")
weighted = f1_score(y_test,y_pred,average="weighted")
print(macro)
print(weighted)
print(micro)
print(accuracy_score(y_test,y_pred))
```

```python
# Report
print(classification_report(y_test,y_pred))
```
