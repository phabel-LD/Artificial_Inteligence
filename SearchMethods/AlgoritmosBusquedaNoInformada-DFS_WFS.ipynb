### *Ejercicio 1:* Aplicar el algoritmo de búsqueda en profundidad para un grafo.

Solución: En clase se discutió la versión iterativea, por lo que para esta versión, se opta por aplicar recursividad en la búsqueda de un nodo y su trayecto desde el nodo raíz.


```python
def busq_prof(grafo, origen = None, nodo_objetivo = None, trayecto = list()):
    # Inicializar nodo origen de no especificarse
    if origen == None:
        origen = next(iter(grafo))
    # Validar nodo origen
    if origen not in grafo:
        print("Nodo origen no válido.")
        return None
    # Empezar trayecto en el nodo origen
    trayecto = trayecto + [origen]
    # Condicion de paro: exito - Nodo objetivo encontrado
    if origen == nodo_objetivo:
        print(f"Nodo objetivo {nodo_objetivo} encontrado en {len(trayecto)-1} pasos en trayecto profundo!")
        return trayecto
    # Expandir a nodos hijos
    for vecino in grafo[origen]:
        # Actualizar trayecto
        if vecino not in trayecto:
            # Recursion a nuevo nodo
            nuevo_trayecto = busq_prof(grafo, vecino, nodo_objetivo, trayecto)
            if nuevo_trayecto:
                return nuevo_trayecto
    #print(f"Nodo objetivo {nodo_objetivo} no encontrado en trayecto: {trayecto}!")
    return None
```

```python
# Grafo de ejemplo (representación de un laberinto)
grafo = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Evaluar funcion de busqueda
print(busq_prof(grafo=grafo, origen='Z', nodo_objetivo='F'))
print("\n\n")

print(busq_prof(grafo=grafo, nodo_objetivo='Z'))
print("\n\n")

print(busq_prof(grafo=grafo, origen='A', nodo_objetivo='A'))
print("\n\n")

print(busq_prof(grafo=grafo, origen='A', nodo_objetivo='F'))
print("\n\n")

print(busq_prof(grafo=grafo, origen='E', nodo_objetivo='F'))
print("\n\n")

print(busq_prof(grafo=grafo, origen='E', nodo_objetivo='A'))
print("\n\n")

print(busq_prof(grafo=grafo, origen='A', nodo_objetivo='C'))
print("\n\n")

print(busq_prof(grafo=grafo, nodo_objetivo='F'))
```

### *Ejercicio 2:* Resolver el 8-puzzle usando la búsqueda en anchura

Solución: Se utilizan funciones para simular movimientos permitidos del 8-puzzle para crear un arbol. Y se procede con la búsqueda en anchura; se usa una cola para recordar los estados visitados.

```python
# Importar librarias requeridas
from collections import deque
import numpy as np
```

```python
# Definir la clase para trabajar con el 8-puzzle
class Puzzle:

    # Constructor
    def __init__(self, tablero):
        # Atributo para el tablero
        self.tablero = tablero
        # Atributo para estado solucion
        self.objetivo = [[1, 2, 3],
                         [4, 5, 6],
                         [7, 8, 0]]
        # Atributo para recordar estados visitados
        self.visitados = set()

    # Mover cero arriba si es posible
    def mover_arriba(self, tablero):
        # Convertir en array de numpy
        matriz = np.array(tablero)
        # Obtener indices del valor '0'
        indice = np.where(matriz == 0)
        # Verificar si el movimiento es posible dado los indices
        if indice[0][0] > 0:
            # Intercambiar valores en las casillas
            c = matriz[indice[0][0] - 1][indice[1][0]]
            matriz[indice[0][0] - 1][indice[1][0]] = matriz[indice[0][0]][indice[1][0]]
            matriz[indice[0][0]][indice[1][0]] = c
        # Regresar matriz/tablero resultante
        return matriz.tolist()

    # Mover cero abajo si es posible
    def mover_abajo(self, tablero):
        # Convertir en array de numpy
        matriz = np.array(tablero)
        # Obtener indices del valor '0'
        indice = np.where(matriz == 0)
        # Verificar si el movimiento es posible dado los indices
        if indice[0][0] < 2:
            # Intercambiar valores en las casillas
            c = matriz[indice[0][0] + 1][indice[1][0]]
            matriz[indice[0][0] + 1][indice[1][0]] = matriz[indice[0][0]][indice[1][0]]
            matriz[indice[0][0]][indice[1][0]] = c
        # Regresar matriz/tablero resultante
        return matriz.tolist()

    # Mover cero a la izquierda si es posible
    def mover_izquierda(self, tablero):
        # Convertir en array de numpy
        matriz = np.array(tablero)
        # Obtener indices del valor '0'
        indice = np.where(matriz == 0)
        # Verificar si el movimiento es posible dado los indices
        if indice[1][0] > 0:
            # Intercambiar valores en las casillas
            c = matriz[indice[0][0]][indice[1][0] - 1]
            matriz[indice[0][0]][indice[1][0] - 1] = matriz[indice[0][0]][indice[1][0]]
            matriz[indice[0][0]][indice[1][0]] = c
        # Regresar matriz/tablero resultante
        return matriz.tolist()

    def mover_derecha(self, tablero):
        # Convertir en array de numpy
        matriz = np.array(tablero)
        # Obtener indices del valor '0'
        indice = np.where(matriz == 0)
        # Verificar si el movimiento es posible dado los indices
        if indice[1][0] < 2:
            # Intercambiar valores en las casillas
            c = matriz[indice[0][0]][indice[1][0] + 1]
            matriz[indice[0][0]][indice[1][0] + 1] = matriz[indice[0][0]][indice[1][0]]
            matriz[indice[0][0]][indice[1][0]] = c
        # Regresar matriz/tablero resultante
        return matriz.tolist()

    # Funcion para obtener todos los estados siguientes dados los 4 movimientos posibles
    def obtener_vecinos(self, nodo):
        # Obtener estados vecinos posibles
        vecinos = list()
        vecinos.append(self.mover_arriba(nodo))
        vecinos.append(self.mover_abajo(nodo))
        vecinos.append(self.mover_izquierda(nodo))
        vecinos.append(self.mover_derecha(nodo))
        # Regresar todos estados nuevos posibles
        return vecinos

    # Funcion para imprimir el camino mas optimo dado el trayecto recordado
    def imprimir_camino(self, camino):
        # Un paso a la vez
        for i, paso in enumerate(camino):
            print(f"Paso {i+1}:")
            # Imprimir en formato de matriz para mas claridad
            for fila in paso:
                print(fila)
            print("\n")

    # Funcion para solucionar el 8-puzzle con busqueda de amplitud
    def solucion_busq_amplitud(self):
        # Lista para recordar los estados visitados
        visitados = []
        # Usar cola para iterar los estados posibles
        trayecto = deque([(self.tablero, [self.tablero])])
        # Repetir mientras haya estados posibles
        while len(trayecto):
            # Tomar la opcion mas a la izquierda por nivel
            nodo, camino = trayecto.popleft()
            # Evaluar solucion
            if np.array_equal(nodo, self.objetivo):
                print("Solucion optima encontrada")
                self.imprimir_camino(camino)
                # Regresar solucion
                return camino
            # Actulizar recorrido y estados visitados
            if tuple(map(tuple, nodo)) not in [tuple(map(tuple, visita)) for visita in visitados]:
                visitados.append(nodo)
                # Obtener los vecinos posibles
                vecinos = self.obtener_vecinos(nodo)
                # Agregar vecinos posibles
                for vecino in vecinos:
                    trayecto.append((vecino, camino + [vecino]))

        # Indicar de no haber solucion
        print("No hay solucion")
        return None
```

```python
# Definir un estado inicial del tablero
estado_inicial = [[4, 1, 3],
                  [7, 6, 0],
                  [5, 2, 8]]

# Instanciar objeto
puzzle_8 = Puzzle(estado_inicial)

# Solucionar 8-puzzle
print(puzzle_8.solucion_busq_amplitud())
```
