# Tensores

Los tensores son objetos matemáticos que, dentro del aprendizaje automático, se utilizan para representar datos numéricos en arreglos. Los tensores se caracterizan por dos elementos:

1. El rango: el cual indica un tipo de objeto tensorial. Quizá la forma más sencilla es entender el rango como el número de índices que tiene un tensor. Si se tiene un sólo índice es de rango 1. Los tensores de rango 2 tendrán 2 índices, etc.
2. Sus dimensiones: Determinan el número de elementos/datos que el tensor puede tener. Por ejemplo, un tensor de rango 1, esto es, un vector, puede tener 3 dimensiones: 3 valores numéricos, o puede tener un millón de dimensiones.

Dentro de los tensores, los elementos más comúnes son los siguientes:

| Rango       | Nombre | Notación |
| ----------- | ----------- | ----------- |
| 0     | Escalar       | $\lambda$ |
| 1   | Vector        | $x_i$ |
| 2   | Matriz        | $A_{i,j}$ |
| 3   | Tensor        | $T_{i,j,k}$ |


## Tensores en Numpy

Numpy nos permite manejar tensores de diferente rango. En este caso, los trata como arreglos (arrays) numéricos. Por lo que para crear un tensor usamos la función <tt>np.array()</tt>. A partir de esta función podemos generar diferentes tipos de tensores.

#### Tensor de rango 1: vector

Un vector es un tensor de rango 1, es decir, tiene un sólo índice:

$$x = x_i$$

Cada entrada $x_i$ del vector contiene información sobre algún dato, por ejemplo un atributo.

```python
import numpy as np

#Tensor de rango 1
x = np.array([2,3,3])

print(x)
```

Los vectores, tanto en 2 como en 3 dimensiones, son visualizables como flechas en espacios de 2 o 3 dimensiones.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, x[0], x[1], x[2])
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])
plt.xlabel('$x_0$')
plt.ylabel('$x_1$')
ax.set_zlabel('$x_2$')
#ax.view_init(30,45)
plt.show()
```

#### Tensor de rango 2: Matriz

Una matriz cuenta con dos índices siendo un tensor de rango 2:

$$A = A_{i,j}$$

Se puede representar como tabla, el primer índice hace referencia a los renglones y el segundo a las columnas:

```python
#Tensor de rango 2 (matriz)
A = np.array([[0, 5, 1],
              [3, 0, 2],
              [0, 2, 0]])

print(A)
```

#### Tensores de mayor rango

Se pueden crear tensores con un rango n mayor a 2; en este caso se manejarán n índices. Por ejemplo un tensor de rango 3 maneja 3 índices. Como se puede observar, el tensor se visualiza como si constara de dos matrices.

```python
#Tensor de rango 3
T = np.array([[[[0, 5, 1],
                [3, 0, 2],
                [0, 2, 0]],
              [[2, 1, 10],
               [5, 4, 9],
               [1, 0, 0.5]]],[[[0, 5, 1],
                [3, 0, 2],
                [0, 2, 0]],
              [[2, 1, 10],
               [5, 4, 9],
               [1, 0, 0.5]]]])

print(T)
print(T.shape)
```

## Tensores en PyTorch y Tensorflow

Paqueterías que serán de suma importancia para el desarrollo de redes neuronales son PyTorch y Tensorflow. En tanto las redes neuronales trabajan con tensores, estas librerías permiten el manejo de diferentes tipos de tensores, de diferente rango. En general, estas librerias crean tensores con diferentes funciones:

1. En pytorch suele usarse <tt>tensor</tt> o <tt>Tensor</tt>.
2. En tensorflow puede usarse <tt>constant</tt>

```python
import torch
import tensorflow as tf

x = torch.tensor([2,3,3])
x_tf = tf.constant([2,3,3])

print(x)
print(x_tf)
```

```python
A = torch.tensor([[0, 5, 1],
                  [3, 0, 2],
                  [0, 2, 0]])

A_tf = tf.constant([[0, 5, 1],
                  [3, 0, 2],
                  [0, 2, 0]])

print(A)
print(A_tf)
```

```python
T = torch.tensor([[[0, 5, 1],
                    [3, 0, 2],
                    [0, 2, 0]],
                  [[2, 1, 10],
                   [5, 4, 9],
                   [1, 0, 0.5]]])

T_tf = tf.constant([[[[0, 5, 1],
                    [3, 0, 2],
                    [0, 2, 0]],
                  [[2, 1, 10],
                   [5, 4, 9],
                   [1, 0, 0.5]]],[[[0, 5, 1],
                    [3, 0, 2],
                    [0, 2, 0]],
                  [[2, 1, 10],
                   [5, 4, 9],
                   [1, 0, 0.5]]]])

print(T)
print(T_tf)
```

### Dimensiones de los tensores

Las dimensiones de los tensores es el número de datos númericos que contiene por cada índice. Por ejemplo, una matriz de $2 \times 3$ nos dice que tiene 2 renglones, es decir que el índice de los renglones son dos. Mientras que tiene 3 índices para las columnas.

Por medios de las paqueterías que estamos revisando, podemos ver las dimensiones que tienen los tensores que hemos creado. En cada paquetería se usan diferentes funciones:

* Numpy: <tt>T.shape</tt>
* Tensorflow: <tt>T.shape</tt>
* PyTorch <tt>T.size()</tt>

```python
print('Tensor {} tiene dimensiones: {}'.format(x, x.size()))
print('Tensor {} tiene dimensiones: {}'.format(A, A.size()))
print('Tensor {} tiene dimensiones: {}'.format(T, T.size()))
```

### Acceder a las entradas del tensor

Acceder a una entrada del tensor implica obtener el dato numérico que se encuentra en el correspondiente índice de ese tensor. Es decir, es obtener el valor $T_{i_1, i_2,...,i_n}$ del tensor.

Por ejemplo, en una matriz queremos ver que valor numérico se encuentra en $A_{1,2}$ es decir el la primera columna y el renglón 2. Para esto, generalmente se usa la notación:

$$T[i_1, i_2, ...,i_n]$$

Con valores específicos para cada índice. Vemos unos ejemplos:

```python
print(A[:,1])
print(T[:,:,0])
```

## Cambio de dimensiones

Las diferentes dimensiones de un tensor pueden ser arregladas de diferente forma. Por ejemplo, las matrices de $n \times m$ son isomorfas a los vectores de dimensión $nm$ por lo que toda matriz puede llevarse a un vector que tenga una dimensión del producto de las dimensiones originales de la matriz. Para realizar esto, PyTorch ofrece dos estrategias:

* <tt>view</tt>: Cambia la forma del tensor usando la misma dirección de memoria.
* <tt>reshape</tt>: Cambia la forma del tensor sin relevancia de la memoria.

```python
T.view(2, 9)
```

```python
A.reshape(9)
```

```python
x = torch.tensor([2.01,3,3])
y = torch.Tensor([2,3,3])
print(y)
print(x.dtype, y.dtype)

torch.dot(x,y)
```

## Tipos de datos

Los tensores pueden contener distintos tipos de datos en un sentido computacional; es decir, pueden contener flotantes, enteros, enteros largos, etc. El tipo de dato de los tensores influye en diferentes formas. Por ejemplo, no se permiten hacer operaciones entre vectores con diferentes tipos de datos; asimismo, las clases generalmente operan con enteros largos <tt>LongTensor</tt> cuando se usan para entrenar un modelo. Los tipos de datos que puede manejar Torch son:

|Tipo de dato | PyTorch     |
| ----------- | ----------- |
| Booleano      | BoolTensor       |
| Flotante 32-bits      | FloatTensor       |
| Flotante 64-bits   | DoubleTensor        |
| Flotante 16-bits   | HalfTensor        |
| Entero 8-bits   | CharTensor        |
| Entero 16-bits   | ShorTensor        |
| Entero 32-bits   | IntTensor        |
| Entero 64-bits   | LongTensor        |

```python
torch.BoolTensor([0,1])
```

```python
a = torch.HalfTensor([0,1])
b = torch.LongTensor([0,1])
c = torch.FloatTensor([0,1])

print(a,b)
print(c)
```

También podemos definir el tipo de dato que contiene el vector a partir de asignarlo con el atributo <tt>dtype</tt>.

```python
a = torch.tensor([0,1], dtype=torch.float32)
print(a)
```

## Propiedades de tensores en PyTorch

Los tensores pueden usarse dentro de diferente forma dentro de un modelo de aprendizaje profundo. Generalmente, los tensores representan los pesos de una red neuronal que buscan optimizarse a partir de un entrenamiento. Sin embargo, puede ser que haya vectores que sean constantes o no se busque actualizar. Para esto, debemos especificar si un tensor requiere o no un gradiente a ser calculado. Para esto PyTorch requiere de especificar si se requerirá o no gradiente de un tensor usando <tt>requires_grad</tt>:

```python
x = torch.FloatTensor([1,3]).requires_grad_(True)
print(x)
```

Si no se especifica este valor, se asumirá que el tensor no requiere gradiente.

El especificar el requerimiento de un gradiente para un tensor permite justamente derivar sobre este tensor cuando se realizan operaciones. Por ejemplo, si elevamos al cuadrado y sumamso, podemos utilizar la función <tt>backward</tt> sobre el resultado, lo que asignará un gradiente al tensor sobre el que se operó.

```python
result1 = x.pow(2)
result = result1.sum()
print(result1)
print(result)

result.backward()
print(x.grad)
```

Podemos hacer operaciones entre tensores constantes (sin un gradiente) y tensores que serán derivados posteriormente. Es importante que cada vez que queramos obtener un gradiente nuevo, reiniciemos el gradiente a 0, pues de lo contrario se sumará el nuevo gradiente al gradiente que tiene el tensor actualmente. Para esto, usamos <tt>grad.zero</tt> que lleva el gradiente del tensor a 0.

```python
a = torch.FloatTensor([5,4])
result = torch.dot(a,x)
#Reinicia el gradiente a 0 para no acumular
x.grad.zero_()
print(result)
result.backward()
print(x.grad)
print(a.grad)
```
