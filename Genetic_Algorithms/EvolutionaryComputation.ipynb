```python
def prob_selection(population: dict()) -> dict():
    data = [value for value in population.values()]
    total = sum(data[i][1] for i in range(len(data)))
    for key in population.keys():
        population[key].append(population[key][1] / total)
    return(population)


population = {
    "011100001": [[2.143, 2.857, 0.714], 4.373],
    "011011001": [[2.143, 2.143, 0.714], 3.279],
    "001011001": [[0.714, 2.143, 0.714], 1.093],
    "001010100": [[0.714, 1.429, 2.857], 2.915]

}

population = prob_selection(population)
population
```
